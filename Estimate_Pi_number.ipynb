```python
import matplotlib.pyplot as plt
import random
import math
```

```python
def estimate_pi_number(n=5):
    points_circle = []
    points_square = []
    num_circle = 0
    num_square = 0
    for _ in range(n):
        x = random.uniform(0,1)
        y = random.uniform(0,1)
        distance = x**2 + y**2
        if distance > 1:
            points_square.append((x,y))
            num_square += 1
        else: 
            points_circle.append((x,y))
            num_circle += 1
    print(num_circle)
    print(num_square)
    return (points_circle, points_square, 4* num_circle / (num_circle + num_square))
```

```python
(points_circle, points_square, pi_value) = estimate_pi_number(9999)
for pC in points_circle:
    plt.plot(pC[0], pC[1], "ro")
for pS in points_square:
    plt.plot(pS[0], pS[1], "go")
print(pi_value)
```
